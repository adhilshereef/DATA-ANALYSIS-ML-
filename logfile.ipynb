```python
file=open(r"/home/cocoslabs/Downloads/log (2).txt","r")
file1=open("/home/cocoslabs/Desktop/log2.txt","w")
content=file.read()

lines=content.split("\n")
delete=['Region Avg IOU:', 'Class:', 'No Obj:','Obj:','Avg Recall:', 'count:']
for line in lines:
    if line.startswith("R"):
        line1=line.split(",")
        for l in line1:
            for z in delete:
                if z in l:
                    l=l.replace(z," ")
            file1.write(l)
        file1.write("\n")
        #print(l,end=" ")
        #print()
file.close()
file1.close()
        
```
